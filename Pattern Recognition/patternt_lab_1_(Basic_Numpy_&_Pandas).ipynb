Sure! Below is a more detailed explanation of how to handle missing values using **Mean**, **Median**, and **Mode**, along with code examples in Python for each method.

---

### **1. Mean (Average)**

- **What is Mean?**
  - The **mean** is the arithmetic average of a set of numbers.
  - It is best used when data is **symmetrical** and has no **outliers**.
  - Formula:  
    \[
    \text{Mean} = \frac{\sum X}{n}
    \]
    where \( X \) is each value in the dataset and \( n \) is the number of values.

- **When to Use:**
  - For **continuous numerical data** like height, weight, temperature, etc.
  - When the data is normally distributed (i.e., symmetric, no extreme outliers).

- **Python Code Example:**

```python
import pandas as pd
import numpy as np

# Sample dataset with missing values (NaN)
data = {'Age': [25, 30, 35, np.nan, 40, np.nan, 50]}
df = pd.DataFrame(data)

# Replace missing values with the mean of the column
mean_value = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_value)

print(df)
```

---

### **2. Median**

- **What is Median?**
  - The **median** is the middle value in a dataset when it is ordered from smallest to largest.
  - If there is an **even number** of values, the median is the average of the two middle numbers.
  
- **When to Use:**
  - When the data contains **outliers** or is **skewed**.
  - The **median** is not affected by extreme values, so it’s more robust than the mean in such cases.
  - Used for both **continuous and ordinal data**.

- **Python Code Example:**

```python
# Replace missing values with the median of the column
median_value = df['Age'].median()
df['Age'] = df['Age'].fillna(median_value)

print(df)
```

---

### **3. Mode**

- **What is Mode?**
  - The **mode** is the value that appears most frequently in a dataset.
  - A dataset may have **no mode**, **one mode** (unimodal), or **multiple modes** (bimodal or multimodal).
  
- **When to Use:**
  - When dealing with **categorical data** or **discrete numerical data**.
  - Especially useful for **nominal** data where there is a strong frequency of certain values.

- **Python Code Example:**

```python
# Sample dataset with categorical data
data = {'Color': ['Red', 'Blue', 'Blue', np.nan, 'Green', np.nan, 'Blue']}
df = pd.DataFrame(data)

# Replace missing values with the mode of the column
mode_value = df['Color'].mode()[0]  # mode() returns a Series, get the first value
df['Color'] = df['Color'].fillna(mode_value)

print(df)
```

---

### **Summary of When to Use Each:**

| Method  | Use Case                                           | Example                                   |
|---------|----------------------------------------------------|-------------------------------------------|
| **Mean** | Best for continuous, symmetrical data without outliers | Heights, weights, temperature            |
| **Median** | Best for skewed data or data with outliers       | Income data, house prices, test scores    |
| **Mode**   | Best for categorical data or data with high frequency values | Favorite colors, shoe sizes, most common category |

---

### **Additional Tips:**
- **Outliers and Mean**: The mean can be heavily affected by outliers (extreme values), making it less reliable when data is skewed or has large outliers.
- **Data Types**: For **numerical columns**, the **mean** and **median** are typically preferred, but for **categorical columns**, the **mode** is most suitable.
- **Mode**: In the case of categorical data with many unique values or no repeated categories, mode might not be effective.
